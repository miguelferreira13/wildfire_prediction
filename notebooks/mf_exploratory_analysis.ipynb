# Exploratory Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('../raw_data/Jan_16/HistoricalWeather.csv')
```

```python
df.Parameter.value_counts()
```

```python
# df_wf[df_wf['Date']=='2014-01-01']
df.info()
```

```python
pivot = df.pivot_table(['min()', 'max()', 'mean()', 'variance()'], ['Date','Region','count()[unit: km^2]'], 'Parameter')
```

```python
pivot
```

```python
pivot.columns = [' '.join(col).strip() for col in pivot.columns.values]
```

```python
pivot.columns
```

```python
pivot = pivot.reset_index()
```

```python
pivot.isna().sum()
```

```python
pivot[["Year", "Month", "Day"]] = pivot["Date"].str.split("-", expand = True)
```

```python
pivot['Date'] = pd.to_datetime(pivot['Date'])
```

```python
pivot.to_csv('../wildfire_prediction/data/weather.csv')
```
